## Dataset Harga Pangan Mingguan Kota Bandung

Data diambil dari website [Pusat Informasi Harga Pangan Strategis Nasional](hargapangan.id) dengan filter Kota Bandung & Tipe Laporan Week to Week setelah itu dilakukan penghapusan row 1 s/d 8, kemudian dilakukan preprocessing awal sbb:

```python
import pandas as pd
```

```python
xls = pd.read_excel("dataset/data_mingguan_harga_pangan_kota_bandung.xls")
```

```python
xls.head()
```

```python
df = xls.copy()
```

```python
df = df.drop(columns="No.")
```

```python
df = df.rename_axis('index',axis=1)
```

```python
df.head()
```

```python
df = df.T.reset_index()
```

```python
df.head()
```

```python
df.columns = df.iloc[0]
```

```python
df = df.drop(df.index[0])
```

```python
df
```

```python
col_names = ['tanggal',
             'beras','beras_kb1','beras_kb2','beras_km1','beras_km2','beras_ks1','beras_ks2', 
             'daging_ayam','daging_ayam_rs',
             'daging_sapi','daging_sapi_k1','daging_sapi_k2',
             'telur_ayam','telur_ayam_rs',
             'bawang_merah','bawang_merah_sedang',
             'bawang_putih','bawang_putih_sedang',
             'cabai_merah','cabai_merah_besar','cabai_merah_keriting',
             'cabai_rawit','cabai_rawit_hijau','cabai_rawit_merah',
             'minyak_goreng','minyak_goreng_curah','minyak_goreng_merk1','minyak_goreng_merk2',
             'gula_pasir','gula_pasir_premium','gula_pasir_lokal'
            ]
```

```python
df.columns = col_names
```

```python
df.tail()
```

```python
df.to_csv('dataset/harga_pangan_mingguan_bandung.csv',index=False)
```
